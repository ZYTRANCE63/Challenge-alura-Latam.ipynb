### Importaci칩n de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. An치lisis de facturaci칩n



```python
import pandas as pd
import numpy as np

# --- 0. PREPARACI칍N Y CONSOLIDACI칍N DE DATOS (Asumido)

# Concatenar todos los DataFrames de las tiendas en un solo DataFrame 'df'
df = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

# Asegurar que las columnas num칠ricas sean de tipo flotante
df['Precio'] = df['Precio'].astype(float)
df['Costo de env칤o'] = df['Costo de env칤o'].astype(float)

# --- C츼LCULO DE M칄TRICAS CLAVE DE FACTURACI칍N ---

# 1. Calcular la Ganancia Bruta
df['Ganancia Bruta'] = df['Precio'] - df['Costo de env칤o']

# 2. Calcular las m칠tricas
total_transacciones = len(df)
ingreso_total = df['Precio'].sum()
ganancia_bruta_total = df['Ganancia Bruta'].sum()
precio_promedio = df['Precio'].mean()
costo_envio_promedio = df['Costo de env칤o'].mean()

# 3. Crear el DataFrame de Resumen
resumen_facturacion = pd.DataFrame({
    'M칠trica de Facturaci칩n': [
        'Total de Transacciones',
        'Ingreso Total',
        'Ganancia Bruta Total',
        'Precio Promedio por Venta',
        'Costo de Env칤o Promedio'
    ],
    'Valor': [
        total_transacciones,
        ingreso_total,
        ganancia_bruta_total,
        precio_promedio,
        costo_envio_promedio
    ]
})

# 4. Formato de salida (m치s legible, con separadores de miles y decimales)
resumen_facturacion['Valor'] = resumen_facturacion['Valor'].apply(
    lambda x: f"${x:,.0f}" if x > 1000 else f"${x:,.2f}"
)

# --- IMPRESI칍N DEL RESULTADO ---
print("## 游눯 An치lisis de Facturaci칩n General")
print(resumen_facturacion.to_markdown(index=False))
```

```python
import pandas as pd
import numpy as np

# --- 0. PREPARACI칍N Y CONSOLIDACI칍N DE DATOS (Asumido) ---
# Se asume que los cuatro archivos se han cargado y combinado en 'df_tiendas' en pasos anteriores.
# Es crucial limpiar los nombres de las columnas y asegurar los tipos de datos.

# 1. Limpieza y c치lculo de Ganancia Bruta (Necesario para c치lculos precisos)
df_tiendas.columns = df_tiendas.columns.str.strip()
df_tiendas['Precio'] = df_tiendas['Precio'].astype(float)
df_tiendas['Costo de env칤o'] = df_tiendas['Costo de env칤o'].astype(float)
df_tiendas['Ganancia Bruta'] = df_tiendas['Precio'] - df_tiendas['Costo de env칤o']

# La columna 'Tienda' ya fue creada y asignada correctamente en 'df_tiendas' previamente.

print("## An치lisis de Ventas Completo en C칩digo Pandas\n")

# --- 1. Facturaci칩n Total de Cada Tienda (Ingresos por Tienda) ---
print("### 1. Facturaci칩n Total (Ingresos) por Tienda\n")

facturacion_por_tienda = df_tiendas.groupby('Tienda')['Precio'].sum().sort_values(ascending=False).reset_index(name='Ingreso Total')

# Formato
facturacion_por_tienda['Ingreso Total'] = facturacion_por_tienda['Ingreso Total'].apply(lambda x: f"${x:,.0f}")

print(facturacion_por_tienda.to_markdown(index=False))
print("-" * 50)


# --- 2. Categor칤as M치s Populares (por Ingreso) ---
print("### 2. Categor칤as M치s Populares (Top 5 por Ingreso)\n")

# Agrupar por categor칤a y sumar el precio
ingresos_por_categoria = df_tiendas.groupby('Categor칤a del Producto')['Precio'].sum().sort_values(ascending=False)

# Calcular el porcentaje del total
total_ingresos = df_tiendas['Precio'].sum()
top_categorias = (
ingresos_por_categoria
                    .head(5)
                    .reset_index())
top_categorias['Porcentaje del Total'] = (top_categorias['Precio'] / total_ingresos) * 100

# Formato y renombramiento de columnas
top_categorias.columns = ['Categor칤a', 'Ingreso Total ($)', 'Porcentaje (%)']

print(top_categorias.to_markdown(index=False, floatfmt=(",.0f", ",.0f", ",.2f")))
print("-" * 50)

# --- 3. Promedio de Calificaci칩n de los Clientes ---
print("### 3. Promedio de Calificaci칩n de los Clientes\n")

calificacion_promedio = df_tiendas['Calificaci칩n'].mean()

print(f"Calificaci칩n Promedio General de la Tienda: {calificacion_promedio:.2f} / 5")
print("-" * 50)

# --- 4. Productos M치s y Menos Vendidos (por Cantidad de Transacciones) ---
print("### 4. Productos M치s y Menos Vendidos\n")

# Conteo de transacciones por producto
conteo_productos = df_tiendas.groupby('Producto').size().reset_index(name='Cantidad de Ventas')
conteo_productos = conteo_productos.sort_values(by='Cantidad de Ventas', ascending=False)


# --- 5. Costo Promedio del Env칤o ---
print("### 5. Costo Promedio del Env칤o\n")

costo_envio_promedio = df_tiendas['Costo de env칤o'].mean()

print(f"Costo Promedio de Env칤o por Transacci칩n: ${costo_envio_promedio:,.2f}")
print("-" * 50)
```

**Reasoning**:
